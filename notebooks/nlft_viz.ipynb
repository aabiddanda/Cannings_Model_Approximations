```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../src/')
import moran_lib as moran
import dtwf_lib as dtwf
import coalescent_lib as coal
%matplotlib inline
```

```python
n = 250
N = 20000
t = int(N/2)
# N = [N] * (t + 1)
ymoran = moran.prob_lineages_step(n,t,N)
ydtwf = dtwf.prob_anc(n, N)

l = len(ymoran)
ymoran = ymoran[1:l]
ydtwf = ydtwf[1:l]
x = [i for i in range(1,n+1)]
# Computing Expectations
nl_moran = sum([i*ymoran[i] for i in range(n)])
nl_dtwf = sum([i*ydtwf[i] for i in range(n)])
print("E(NLFT | Moran) = %0.8f\nE(NLFT | DTWF) = %0.8f" % (nl_moran, nl_dtwf))

fig, (ax1, ax2) = plt.subplots(2, 1, sharey=True, sharex=True)
ax1.bar(x,ymoran, linewidth=0, color='blue', align='center')
ax2.bar(x,ydtwf, linewidth=0, color='blue', align='center')
ax1.set_title('Moran Model', fontsize=14)
ax2.set_title('DTWF', fontsize=14)
ax1.set_ylabel(r'Probability', fontsize=14)
ax2.set_ylabel(r'Probability', fontsize=14)
# ax1.set_xlabel(r'$m$')
ax2.set_xlabel(r'$m$')
ax1.set_xlim(n-10,n+1)
plt.tight_layout();
```
